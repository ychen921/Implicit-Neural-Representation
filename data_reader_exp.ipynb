```python
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from data_reader import SingleImageDataset
import numpy as np
```

```python
img_path = r'data/'
dataset = SingleImageDataset(img_path + 'image.png')
```

```python
# Initializing a DataLoader
batch_size = 1
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

# Using dataloader to retrieve, reassemble, and display image

reasmb_img = np.zeros((dataset.image.permute(1, 2, 0).numpy().shape))

for batch in dataloader:
    x, y, intensity = batch["x"], batch["y"], batch["intensity"]
    reasmb_img[y.item(), x.item()] = np.asarray(intensity)[0]*255

plt.imshow(reasmb_img.astype(np.uint8))
plt.axis('off')
plt.show()
```
